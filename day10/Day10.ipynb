```python
import sys
import numpy as np
from functools import lru_cache

def read_file(file_name):
    with open(file_name) as input_file:
        return input_file.read().splitlines()

def is_open(char):
    return '<' in char or '{' in char or '[' in char or '(' in char

def is_closed(char):
    return '>' in char or '}' in char or ']' in char or ')' in char

def matching(Open,char):
    s=Open[len(Open)-1];
    return (s=='<' and char=='>') or (s=='[' and char==']') or (s=='(' and char==')') or (s=='{' and char=='}') 
    
def value(char):
    if '[' in char:
        return 2;
    if '(' in char:
        return 1;
    if '<' in char:
        return 4;
    if '{' in char:
        return 3;
    
def complete(line):
    Open=[];
    for char in line:
           if is_open(char):
                Open.append(char);
           if is_closed(char):
                if matching(Open,char):
                    Open.pop();
                else:
                    return {},False
    return Open,True;

def score(List):
    score=0;
    for char in List:
        #print(char,value(char))
        score *= 5
        score += value(char)
    return score
    

if __name__ == '__main__':
    #read lines
    all_lines=read_file('input')
    #all_lines=read_file('input-test')

    Scores=[];
    
    #make a matrix out of the list
    for line in all_lines:
        List,Bool=complete(line)
        
        if Bool:
            Scores.append(score(List[::-1]))
    print(len(Scores),sorted(Scores),sorted(Scores)[int(len(Scores)/2)])
    
```
