```python
import sys
import numpy as np
import copy
from functools import lru_cache,wraps
from operator import itemgetter

import collections
def read_file(file_name):
    with open(file_name) as input_file:
        list_of_lines=input_file.read().splitlines()
    Numbers=[]
    for lines in list_of_lines:
        input=[char for char in lines]
        n=0
        line=[]
        for i in input:
            if '[' in i:
                n+=1
            elif ']' in i:
                n-=1
            elif i.isnumeric():
                line.append([int(i),n])
        Numbers.append(line)
    return Numbers

def reduce(num):
    number=num.copy()
    max_level=max(number,key=itemgetter(1))[1]
    max_value=max(number,key=itemgetter(0))[0]
    #print('max level {} max value {}' .format(max_level,max_value))
    while(max_level>4 or max_value>9):
        if(max_level>4):
            number=explode(number)
            max_level=max(number,key=itemgetter(1))[1]
            max_value=max(number,key=itemgetter(0))[0]
        elif(max_value>9):
            number=split(number)
            max_level=max(number,key=itemgetter(1))[1]
            max_value=max(number,key=itemgetter(0))[0]
    return number

def explode(num):
    number=num.copy()
    index=next(i for i,v in enumerate(number) if v[1]>4)
    #print('explode at index {}'.format(index))
    if index>0:
        number[index-1][0]+=number[index][0]
    if index<len(number)-2:
        number[index+2][0]+=number[index+1][0]
    number[index][0]=0;
    number[index][1]-=1;
    number.pop(index+1)
    #print(number)
    return number
    
def split(num):
    number=num.copy()
    index=next(i for i,v in enumerate(number) if v[0]>9)
    #print('split at index {}'.format(index))
    value=number[index][0]
    level=number[index][1]
    low_half=value//2
    up_half=-(-value//2)
    number.insert(index,[low_half,level+1])
    number[index+1][0]=up_half
    number[index+1][1]+=1
    #print(number)
    return number
    
    
def Sum(num1,num2):
    number1=copy.deepcopy(num1)
    number2=copy.deepcopy(num2)
    #print('sum')
    #print(number1)
    #print(number2)
    for i in number2:
        number1.append(i)
    for i in number1:
        i[1]+=1
    return number1                    
    
def magnitude(num):
    number=num.copy()
    max_level=max(number,key=itemgetter(1))[1]
    while(max_level>1):
        index=next(i for i,v in enumerate(number) if v[1]==max_level)
        #print('index of max level {}'.format(index))
        number[index][0]  = number[index][0]*3+number[index+1][0]*2
        number[index][1] -= 1
        number.pop(index+1)
        #print(number)
        max_level=max(number,key=itemgetter(1))[1]
    index=0
    return number[index][0]*3+number[index+1][0]*2
    
if __name__ == '__main__':
        
    Numbers = read_file('input')
    for n in range(0,len(Numbers)):
        Numbers[n]=reduce(Numbers[n])
    print(Numbers)

    Result=Numbers[0]
    
    for n in range(1,len(Numbers)):
        #print(n)
        #print(Numbers[0])
        number=copy.deepcopy(Numbers[n])
        temp_sum=Sum(Result,number)
        #print(Numbers[0])
        Result=reduce(temp_sum)
        #print(Numbers[0])
        
    print(magnitude(reduce(Result)))
    #Lx,Ly,Cave = read_file('input')
   
    Max=0;
    for n in range(0,len(Numbers)):
        for p in range(n+1,len(Numbers)):
                #print(Numbers[n])
                #print(Numbers[p])
                Result=reduce(Sum(Numbers[n],Numbers[p]))
                #print(Result)
                sumnp=magnitude(Result)
                sumpn=magnitude(reduce(Sum(Numbers[p],Numbers[n])))
                Max=max([Max,sumnp,sumpn])
    print(Max)
```
