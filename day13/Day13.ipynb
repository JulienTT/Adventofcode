```python
import sys
import numpy as np
def read_file(file_name):
    with open(file_name) as input_file:
        list_of_lines=input_file.read().splitlines()
    list_of_fold=[line.split(sep=' ') for line in list_of_lines if len(line.split(sep=' '))>1]
    list_of_coordinates=[line.split(sep=',') for line in list_of_lines if len(line.split(sep=','))>1]   
    return list_of_fold,list_of_coordinates

def fold_cave(cave,fold,xm,ym):
    if fold[0]==0:
        for i in range(fold[1]+1,xm+1):
            if 2*fold[1]-i>=0:
                for j in range(0,ym+1):
                    cave[j,2*fold[1]-i]+=cave[j,i]
        return fold[1]-1,ym
    if fold[0]==1:
        for i in range(fold[1]+1,ym+1):
            if 2*fold[1]-i>=0:
                for j in range(0,xm+1):
                    cave[2*fold[1]-i,j]+=cave[i,j]
        return xm,fold[1]-1
    
def count_cave(cave):
    return np.count_nonzero(cave[0:ym+1,0:xm+1], axis=None)

if __name__ == '__main__':
   
    list_of_fold,list_of_coordinates = read_file('input')
    
    xm=max([int(i[0]) for i in list_of_coordinates])
    ym=max([int(i[1]) for i in list_of_coordinates])
    
    cave=np.zeros((ym+1,xm+1),np.int64)
    for line in list_of_coordinates:
        cave[int(line[1]),int(line[0])]=1
        
    folds=[]
    for line in list_of_fold:
        if 'y' in line[2]:
            yf=int(line[2].split(sep='=')[1])
            folds.append([1,yf])
        if 'x' in line[2]:
            xf=int(line[2].split(sep='=')[1])
            folds.append([0,xf])
    #print(folds)
    

    for fold in folds:
        xm,ym=fold_cave(cave,fold,xm,ym)
    String=[]
    
    for i in range(0,ym+1):
        String.append('')
        for j in range(0,xm+1):
            if(cave[i,j]>1):
                cave[i,j]=1
            if (cave[i,j]==1):    
                String[i] += str(cave[i,j])
            else:
                String[i] += '.'
            
    for i in String:
        print(i)
```
