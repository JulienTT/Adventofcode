```python
import sys
import numpy as np
def read_file(file_name):
    with open(file_name) as input_file:
        return input_file.read().splitlines()

def process_lines(all_lines):
    lines=[line.split(sep='|') for line in all_lines]
    _input=[]
    _output=[]
    for line in lines:
        _input.append(line[0])
        _output.append(line[1])
    return _input,_output

def number_of_unique(entry):
    array_of_length=[len(i) for i in entry.split(sep=' ')]
    return [int(np.count_nonzero(line==2 or line==4 or line==3 or line==7)) for line in array_of_length]
    
if __name__ == '__main__':
    #read lines
    #all_lines=read_file('input-test')
    all_lines=read_file('input')

    _input,_output = process_lines(all_lines)
    
    print(sum([sum(i) for i in [number_of_unique(line) for line in _output]]))
   

    
```

```python
#return a list of words of given size in words
def by_size(words, size):
    return [word for word in words if len(word) == size]

def contains(list1,list2):
    #letters contains False if a character of list 2 is not in list 1
    letters=[char in list1 for char in list2]
    #if only Trues, then list2 is in list1
    return False not in letters

def find_entries(line_in):
        entries={}
        # 1 is only segment with 2 letters
        entries[1]=by_size(line_in,2)[0]
        # 4 is only segment with 4 letters
        entries[4]=by_size(line_in,4)[0]
        # 7 is only segment with 3 letters
        entries[7]=by_size(line_in,3)[0]
        # 8 is only segment with 8 letters
        entries[8]=by_size(line_in,7)[0]
        # get six-letter segments
        six_letters=by_size(line_in,6)
        # get five-letter segments
        five_letters=by_size(line_in,5)
       
        # elements with 6 letters: 0 6 9
        for entry in six_letters:
            # 9 is only element with 4 in it
            if contains(entry,entries[4]):
                entries[9]=entry
            # 6 is only element without 1
            elif not contains(entry,entries[1]):
                entries[6]=entry
            # 0 is the only element that fails both tests
            else:
                entries[0]=entry

        # elements with 5 letters: 2 3 5
        for entry in five_letters:
            # 3 is only sedment with 1 in it
            if contains(entry,entries[1]):
                entries[3]=entry
            # 2 is the only element not contained in 9
            if not contains(entries[9],entry):
                entries[2]=entry
        #5 is the only element left over
        five_letters.remove(entries[3])
        five_letters.remove(entries[2])
        entries[5]=five_letters[0]
        return entries

def number_of_output(out_in,entries):
    string=''
    for n in range(1,len(out_in)):
        #for each entry of the output
        i=out_in[n]            
        #find the entries to which it corresponds
        for j in entries:
            if(sorted(i)==sorted(entries[j])):
                string=string+str(j)
    return string

    
if __name__ == '__main__':
    #read lines
    all_lines=read_file('input')
    #all_lines=read_file('input-test')
    #all_lines=read_file('input-shortest')

    _input,_output = process_lines(all_lines)
    
    Sum=0
    # we treat the entries line by line
    for i in range(0,len(_input)):
        # make a list of entries
        line_in=_input[i].split(sep=' ')
        
        # find the correct entries
        entries=find_entries(line_in)
        
        # make a list of output
        out_in=_output[i].split(sep=' ')
        
        # get a number out of it
        string=number_of_output(out_in,entries)
        
        Sum+=int(string)
    print(Sum)
```
