```python
import numpy as np
from collections import Counter
# open the file
f = open('input', 'r')

#create a list with each line as an element
words = f.read().splitlines() 

#close the file
f.close()

#the number of lines is given by the length of the list words.
nbline=len(words)

#the number of bits in each entry is given by the length of an element of words
nbletters=len(words[0])

#create a null array that will be populated with the elements of the list words
data=np.zeros(shape=(nbline,nbletters))

#populate. Surely there must be a better way...
i=0
for c in words:
    j=0
    for letters in list(c):
        data[i,j]=letters
        j+=1;
    i+=1;
      
newdata_g=data
j=0;
#as long as there is more than one line left in the array
while(len(newdata_g)>1):
    # find the most and second most common bit at the place j
    a=Counter(newdata_g[:,j]).most_common(2)
    
    # if two bits are equally common, choose "1"
    if a[0][1]==a[1][1] :
        most_common=1
    # else, choose the most common bit
    else:
        most_common=int(a[0][0])
    
    #rebuild the array keeping only the lines where the j^th term is equal to most common
    newdata_g = newdata_g[ newdata_g[:,j]==most_common , : ]

    # iterate with the next bit
    j += 1;
    
newdata_e=data
j=0;
#as long as there is more than one line left in the array
while(len(newdata_e)>1):
    # find the most and second most common bit at the place j
    a=Counter(newdata_e[:,j]).most_common(2)
    # if two bits are equally common, choose "0"
    if a[0][1]==a[1][1] :
        most_common=0
    # else, choose the most common bit
    else:
        most_common=int(a[1][0])
    #rebuild the array keeping only the lines where the j^th term is equal to second most common
    newdata_e = newdata_e[ newdata_e[:,j]==most_common , : ]
    #iterate with the next bit
    j += 1;

#get the first elements of the arrays
newdata_e=newdata_e[0]    
newdata_g=newdata_g[0]    

#convert them in binaries
gamma=0
eps=0
L=nbletters-1
for j in range(nbletters):
    gamma += newdata_g[L-j]*2.**j
    eps += newdata_e[L-j]*2.**j

print(int(gamma*eps))
```
